```python
import numpy as np
from math import e, pi
import matplotlib.pyplot as plt
%matplotlib inline
```

## Formula

```python
def euler(phi):
    return np.cos(phi)+1j*np.sin(phi)

def euler2(phi, amp=1):
    return amp*np.exp(1j*phi)
```

```python
euler(1)
```

```python
euler2(1)
```

```python
phis=np.linspace(0, 2*np.pi,num=40)
phis
```

```python
ys=np.array([euler2(phi) for phi in phis])
```

```python
ys
```

```python
plt.plot(phis, ys)
```
